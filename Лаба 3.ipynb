```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

data={ 1650: 700,
     1718: 5600,
     1811: 9200,
     1840: 16500,
     1856: 12900,
     1886: 27000,
     1913: 100700,
     1914: 99900,
     1920: 92800,
     1926: 97737,
     1933: 167900, 
     1939: 250011,
     1945: 264800,
     1959: 546878,
     1967: 704000,
     1973: 844000, 
     1979: 969289,
     1985: 1040000,
      1987:1092000,
     1989: 1082052,
      1991: 1097000,
      1993: 1096000,
      1996: 1094000,
      1999: 1088900,
      2001: 1088800,
      2002: 1042437,
      2005: 1036026,
      2007: 1022575,
      2009: 1024842,
      2012: 1072291,
      2015: 1105667,
      2018: 1120547,
      2020: 1128787,
      2021: 1144809
}
years=data.keys()
people=data.values()
years=np.array(list(years))
people=np.array(list(people))

def dydt(y, t):
    k=0.01525
    return k*y
solve=odeint(dydt,y0=4000,t=years)
solve=solve.flatten()
%matplotlib inline
plt.plot(years, solve, label = r"Жесткая модель Мальтуса", color="red")
plt.plot(years, people, label = r"Население Уфы", color="green")
plt.grid(True)
plt.legend()
plt.show()

```
